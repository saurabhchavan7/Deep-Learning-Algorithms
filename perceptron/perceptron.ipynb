```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('placement.csv')
```

```python

print(df.shape)
df.head()
```

```python
sns.scatterplot(x='cgpa', y='resume_score', hue='placed', data=df)
plt.title('CGPA vs Resume Score')
plt.xlabel('CGPA')
```

- our data seems to be linearly seperable hence we are good to go
- will apply perceptron on this and classify in 2 regions

```python
# will extract x and y
X = df.iloc[:,0:2]
y = df.iloc[:,-1]
```

```python
from sklearn.linear_model import Perceptron

# creeated an object p # we will use this object to fit the data
p = Perceptron()
```

```python
# calling p.fit and passing X and y and training the model
p.fit(X,y)
```

```python
# p has the attributes called coefficients which are the weights of the perceptron
p.coef_
```

```python
p.intercept_
# this is the bias term of the perceptron that is b 
```

```python
from mlxtend.plotting import plot_decision_regions
# we can use this to plot the decision boundary 
```

```python
plot_decision_regions(X.values, y.values, clf=p, legend=2)
plt.xlabel('CGPA')
plt.ylabel('Resume Score')
```
