```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
import seaborn as sns
from mlxtend.plotting import plot_decision_regions

import tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout
from tensorflow.keras.optimizers import Adam
```

```python
X, y = make_moons(100, noise=0.25,random_state=2)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:,0], X[:,1], c=y)
plt.show()
```

```python
model1 = Sequential()

model1.add(Dense(128,input_dim=2, activation="relu"))
model1.add(Dense(128, activation="relu"))
model1.add(Dense(1,activation='sigmoid'))

model1.summary()
```

```python
adam = Adam(learning_rate=0.01)
model1.compile(loss='binary_crossentropy', optimizer=adam, metrics=['accuracy'])

history1 = model1.fit(X, y, epochs=2000, validation_split = 0.2,verbose=0)
```

```python
# so we can see we have deliberately set epochs to a high number
# so this will become the case of overfitting
```

```python
plot_decision_regions(X, y.astype('int'), clf=model1, legend=2)
plt.xlim(-2,3)
plt.ylim(-1.5,2)
plt.show()
```

```python
plt.plot(history1.history['loss'])
plt.plot(history1.history['val_loss'])
```

```python
# so what will do next is will add regularization to the model
# so we have added kernel regularization to the model 
# we have l2 regularization
# and we have set the regularization parameter to 0.01
```

```python
model2 = Sequential()

model2.add(Dense(128,input_dim=2, activation="relu",kernel_regularizer=tensorflow.keras.regularizers.l1(0.001)))
model2.add(Dense(128, activation="relu",kernel_regularizer=tensorflow.keras.regularizers.l1(0.001)))
model2.add(Dense(1,activation='sigmoid'))

model2.summary()
```

```python
adam = Adam(learning_rate=0.01)
model2.compile(loss='binary_crossentropy', optimizer=adam, metrics=['accuracy'])

history2 = model2.fit(X, y, epochs=2000, validation_split = 0.2,verbose=0)
```

```python
plot_decision_regions(X, y.astype('int'), clf=model2, legend=2)
plt.xlim(-2,3)
plt.ylim(-1.5,2)
plt.show()
```

```python
# so this is a much cleaner decision boundary
# this will perform better on the test set
```

```python
plt.plot(history2.history['loss'])
plt.plot(history2.history['val_loss'])
```

```python
model1_weight_layer1 = model1.get_weights()[0].reshape(256)
model2_weight_layer1 = model2.get_weights()[0].reshape(256)
```

We use `model1.get_weights()[0].reshape(256)` and `model2.get_weights()[0].reshape(256)` to extract and flatten the weights from the first layer of each model. 

By plotting these weights with a box plot, we can visually compare the distribution and range of weights for the model without regularization (`model1`) and the model with regularization (`model2`). 

Regularization (like L1 or L2) helps prevent overfitting by keeping the weights smaller and more evenly distributed. So, the box plot will show that the weights in the regularized model (`model2`) are generally closer to zero and have less spread compared to the non-regularized model (`model1`). This makes it easy to see the effect of regularization on the model's weights.

A PDF (Probability Density Function) plot helps us understand the distribution of the weights in each model's first layer. By visualizing the PDF, we can see how spread out or concentrated the weights are. 

For example, if the weights are tightly clustered around zero, it suggests that regularization is effectively constraining the model, reducing overfitting. If the weights are more spread out, it may indicate less regularization and a higher risk of overfitting. Comparing the PDF plots for `model1` (without regularization) and `model2` (with regularization) makes it easy to see the impact of regularization on the weight distributions.
